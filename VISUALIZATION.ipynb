```python
subjects = ("Maths","DBMS","OS","ML","Python")
avg_marks = (78,72,69,63,85)
plt.figure(figsize=(6,4))
plt.bar(subjects,avg_marks)
plt.title("average marks per subject")
plt.xlabel("subject")
plt.ylabel("Average Marks")
plt.ylim(0,100)
plt.grid(axis="y",linestyle="--",alpha=0.4)
plt.show
```

```python
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
```

```python
hours = np.array([1,2,3,4,5,6,7,8]).reshape(-1,1) 
marks = np.array([35,40,50,55,65,70,75,85]) 
plt.figure(figsize=(6,4))
plt.scatter(hours,marks)
plt.title("Hours Studied vs Marks")
plt.xlabel("Hours Studied")
plt.ylabel("Marks")
plt.grid(True, linestyle="--", alpha=0.4)
plt.show()
```

```python
model = LinearRegression() 
model.fit(hours,marks)
print("slope(m):",model.coef_[0])
print("Intercept(b):",model.intercept_)
```

```python
hours_new = np.linspace(1,10,100).reshape(-1,1)
marks_pred = model.predict(hours_new)

plt.figure(figsize=(10,6))
plt.scatter(hours,marks,color="blue",label="Actual Data")
plt.plot(hours_new,marks_pred,color="red",label="Fitted Line")
plt.title("Hours vs Marks")
plt.xlabel("Hours")
plt.ylabel("Marks")
plt.grid(True,linestyle="--",alpha=0.4,color="red")
plt.legend()
plt.show()
```

```python
my_hours = np.array([[5]])
my_pred = model.predict(my_hours)[0]
print(f"If you study {my_hours[0][0]} hours, model predict ~{my_pred:.2f} marks")
```
